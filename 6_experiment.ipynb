```python
import pandas as pd
import matplotlib.pyplot as plt
from os import path
import numpy as np
import seaborn as sns
sns.set()
```

```python
df_motion = pd.read_csv(path.join('data','6_slow.csv'))
df_sensor = pd.read_csv(path.join('data', '6s_slow.csv'))
```

```python
plt.plot(df_sensor.Time, df_sensor.Relative_Angle, 'o')
```

```python
plt.plot( df_motion.t, df_motion.theta, 'r.')
```

```python
plt.plot( df_motion.t, df_motion.theta, 'r.' )
plt.plot(df_sensor.Time, df_sensor.Relative_Angle, 'o')
```

```python
plt.figure(2)
plt.plot(df_motion.t, df_motion.theta, 'r.', label='Motion')
plt.plot(df_sensor.Time+.15,df_sensor.Relative_Angle , 'b.', label='Sensor')
plt.title('Angle vs Time')
plt.ylabel('Angle [Degrees]')
plt.xlabel('Time [s]')
plt.legend(loc='upper left')
plt.savefig("4_rpm_cmp.png", dpi=200)
```
